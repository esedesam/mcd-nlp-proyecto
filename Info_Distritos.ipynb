## Incorporación de información inicial

Web de la información: https://alegria-realestate.com/es/articles/the-city-of-valencia

```python
# Función para extraer información de una página web
def extract_information(url):
    # Enviar una solicitud GET a la URL
    response = requests.get(url)
    
    # Analizar el contenido HTML
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Inicializar variables para almacenar el distrito e información
    distrito = ''
    informacion = ''
    
    # Inicializar una lista para almacenar los datos
    datos = []
    
    # Inicializar una bandera para saltar el primer párrafo
    saltar_primer_parrafo = True
    
    # Recorrer todos los elementos en la página
    for elemento in soup.find_all(['h2', 'p', 'li']):
        # Si el elemento es un encabezado, actualizar el distrito
        if elemento.name == 'h2':
            # Si había alguna información previa, almacenarla
            if distrito and informacion:
                datos.append({'Distrito': distrito, 'Información': informacion})
                informacion = ''  # Reiniciar la información
            distrito = elemento.text  # Actualizar el distrito
        # Si el elemento es un párrafo o un elemento de lista y no es el primero, agregar su texto a la información
        elif elemento.name in ['p', 'li'] and not saltar_primer_parrafo:
            # Agregar el texto si no contiene "en el mapa"
            if "en el mapa" not in elemento.text:
                informacion += elemento.text + '\n'  # Agregar el texto
        # Establecer la bandera en Falso después del primer párrafo
        if elemento.name == 'p' and saltar_primer_parrafo:
            saltar_primer_parrafo = False
    
    # Agregar la última pieza de información
    if distrito and informacion:
        datos.append({'Distrito': distrito, 'Información': informacion})
    
    return datos

# URL de la página web para hacer scraping
url = 'https://alegria-realestate.com/es/articles/the-city-of-valencia'

# Extraer información de la página web
datos = extract_information(url)

# Crear un DataFrame
df = pd.DataFrame(datos)

# Imprimir el DataFrame
df
```

```python
# Vamos a quedarnos solo con el nombre del distrito

df = pd.DataFrame(datos)
df['Distrito'] = df['Distrito'].apply(lambda x: x.split('–')[-1].strip())

df
```

```python
df = df.loc[0:19]

# Función para extraer barrios y actualizar el DataFrame
def extract_neighborhoods(df):
    # Crear una nueva columna para los barrios
    df['Barrios'] = ''

    # Iterar a través de cada fila en el DataFrame
    for index, row in df.iterrows():
        # Dividir la columna 'Información' por el carácter de nueva línea
        partes_info = row['Información'].split('\n')

        # Extraer la línea que contiene los barrios
        for parte in partes_info:
            if parte.startswith('Barrios:'):
                barrios = parte.replace('Barrios:', '').strip()
                # Actualizar la columna 'Barrios'
                df.at[index, 'Barrios'] = barrios
                # Eliminar la línea de barrios de 'Información'
                row['Información'] = row['Información'].replace(parte, '').strip()

    return df

# Extraer barrios y actualizar el DataFrame
df = extract_neighborhoods(df)

# Imprimir el DataFrame actualizado
df
```

```python
# Ajustes manuales de Campanar (No sigue el formato de los demás distritos)

df["Distrito"][3] = "Campanar" # Cambio de nombre
df["Barrios"][3] = "Campanar, Les Tendetes, El Calvari y Sant Pau." # Agrrgar barrios


df['Información'][3] = df['Información'][3][48:]
df['Información'][3] = df['Información'][3].replace('\n', ' ') # 
# Aquí eliminamos los barrios de la información y eliminamos los caracteres \n


df
```

## Incorporación de información adicional para algunos distritos

```python
# Send a GET request to the URL
url = "https://valenciasecreta.com/mejores-barrios-valencia/"
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all the <h2> tags containing neighborhood names
    neighborhood_sections = soup.find_all("h2")

    # Initialize lists to store neighborhood names and information
    neighborhood_names = []
    neighborhood_info = []

    # Iterate through the neighborhood sections and extract their names and corresponding info
    for i in range(len(neighborhood_sections)):
        # Get the name of the neighborhood
        neighborhood_name = neighborhood_sections[i].text.strip()
        
        # Initialize a string to store the information about the neighborhood
        info = ""
        
        # Find all the <p> tags containing information about the neighborhood until the next <h2> tag
        next_tag = neighborhood_sections[i].find_next_sibling()
        while next_tag and next_tag.name != "h2":
            if next_tag.name == "p":
                info += next_tag.text.strip() + "\n"  # Append the text of the <p> tag to the info string
            next_tag = next_tag.find_next_sibling()
        
        # Append the neighborhood name and information to the lists
        neighborhood_names.append(neighborhood_name)
        neighborhood_info.append(info.strip())  # Remove leading and trailing whitespace from the info string

    # Create a DataFrame from the lists
    df = pd.DataFrame({"Neighborhood": neighborhood_names, "Information": neighborhood_info})
else:
    print("Failed to retrieve the webpage.")
```

Cogemos más información de otros barrios

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Obtener el HTML de la URL
url = "https://blog.tiko.es/mejores-barrios-valencia/"
response = requests.get(url)
html = response.content

# Analizar el HTML
soup = BeautifulSoup(html, "html.parser")

# Encontrar todos los elementos h3 (nombres de los barrios)
barrios = soup.find_all("h3")

# Crear listas para almacenar los nombres de los barrios y su información
barrios_list = []
info_list = []

# Iterar sobre los elementos h3 para obtener la información de cada barrio
for i in range(len(barrios)):
    # Nombre del barrio
    barrio = barrios[i].text.strip()
    barrios_list.append(barrio)
    
    # Obtener todos los elementos hermanos del barrio hasta el siguiente barrio
    info = ""
    next_sibling = barrios[i].find_next_sibling()
    while next_sibling and next_sibling.name != "h3":
        info += next_sibling.text.strip() + " "
        next_sibling = next_sibling.find_next_sibling()
    info = info.split(barrios[i + 1].text.strip())[0] if i < len(barrios) - 1 else info
    info_list.append(info.strip())

# Crear el DataFrame
data = {"Barrios": barrios_list, "Información": info_list}
df = pd.DataFrame(data)

# Imprimir el DataFrame
df
```
